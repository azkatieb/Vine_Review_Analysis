```python
# Add the Pandas dependency.
import pandas as pd
import os
```

```python
# File to load
vine_csv= "vine_table1.csv"
```

```python
# read vine table csv
vine_table_df = pd.read_csv(vine_csv)
vine_table_df
```

```python
vine_table_df.dtypes
```

```python
helpful_reviews_df = vine_table_df[vine_table_df['total_votes'] >= 20]
helpful_reviews_df
```

```python
helpful_reviews_df.dtypes
```

```python
reliable_votes = helpful_reviews_df.loc[(helpful_reviews_df['helpful_votes']/helpful_reviews_df['total_votes'] >= 0.5)]
reliable_votes
```

```python
paid_reviews_df = reliable_votes.loc[(reliable_votes['vine'] == "Y")]
paid_reviews_df.count()
```

```python
unpaid_reviews_df = reliable_votes.loc[(reliable_votes['vine'] == "N")]
unpaid_reviews_df
```

```python
total_reviews_df = reliable_votes.count()
print (total_reviews_df)
```

```python
five_star_df = reliable_votes.loc[(reliable_votes['star_rating'] == "5")]
five_star_df.count()
```

```python
paid_fivestar = five_star_df.loc[(five_star_df['vine'] == "Y")]
paid_fivestar.count()
```

```python
five_star_all = five_star_df.count()
vine_five = paid_fivestar.count()
perc_paid = vine_five/five_star_all*100
perc_paid
```

```python
unpaid_fivestar = five_star_df.loc[(five_star_df['vine'] == "N")]
unpaid_fivestar.count()
```

```python
five_star_all = five_star_df.count()
novine_five = unpaid_fivestar.count()
perc_unpaid = novine_five/five_star_all*100
perc_unpaid
```
